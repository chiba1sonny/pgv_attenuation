```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
r=np.arange(3,100,0.01)
```

```python
PGV6=10**(0.69*6+0.0042*8-1.75-np.log10(r+0.0033*10**(0.5*6))-0.002*r)
PGV65=10**(0.69*6.5+0.0042*8-1.75-np.log10(r+0.0033*10**(0.5*6.5))-0.002*r)
PGV7=10**(0.69*7+0.0042*8-1.75-np.log10(r+0.0033*10**(0.5*7))-0.002*r)
PGV75=10**(0.69*7.5+0.0042*8-1.75-np.log10(r+0.0033*10**(0.5*7.5))-0.002*r)
PGV8=10**(0.69*8+0.0042*8-1.75-np.log10(r+0.0033*10**(0.5*8))-0.002*r)
```

```python
plt.plot(r,PGV6, label='Mw=6.0',color='gray')
plt.loglog(r,PGV6,color='gray')

plt.plot(r,PGV65, label='Mw=6.5',color='pink')
plt.loglog(r,PGV65,color='pink')

plt.plot(r,PGV7, label='Mw=7.0',color='red')
plt.loglog(r,PGV7,color='red')

plt.plot(r,PGV75, label='Mw=7.5',color='green')
plt.loglog(r,PGV75,color='green')

plt.plot(r,PGV8, label='Mw=8.0',color='blue')
plt.loglog(r,PGV8,color='blue')

plt.xlim(1,100)
plt.ylim(1,200)

plt.xlabel('Fault distance km')
plt.ylabel('PGV cm/s')

plt.title('predicted attenuation curves(Depth=8km)')

plt.legend()
plt.show()
```

# kumamoto0414

```python
import pandas as pd
```

```python
kumamoto0414=pd.read_csv('kumamoto.csv')
```

```python
kumamoto0414.head()
```

```python
M=kumamoto0414.iloc[:,0].values
Rkumamoto0414=kumamoto0414.iloc[:,1].values
D=kumamoto0414.iloc[:,2].values
```

```python
Ypgvkumamoto0414=10**(0.69*M+0.0042*D-1.75-np.log10(Rkumamoto0414+0.0033*10**(0.5*M))-0.002*Rkumamoto0414)
YYpgvkumamoto0414=10**(0.58*M+0.0038*D-1.29-np.log10(Rkumamoto0414+0.0028*10**(0.5*M))-0.002*Rkumamoto0414)
```

```python
Kpgvkumamoto0414=pd.read_csv('pgvkumamoto.csv')
```

```python
plt.title('kumamoto0414') 
plt.scatter(Rkumamoto0414,Kpgvkumamoto0414,s=13,color='green',label='kansoku',alpha=1) 
plt.scatter(Rkumamoto0414,Ypgvkumamoto0414,s=13,color='red',label='yosoku',alpha=1)
plt.scatter(Rkumamoto0414,YYpgvkumamoto0414,s=13,color='gray',label='SM',alpha=1)

plt.xlim(10,500) 
plt.ylim(0.1,10000) 
plt.xlabel('distance(km)') 
plt.ylabel('pgv(cm/s)') 
plt.yscale('log') 
plt.xscale('log') 
plt.legend() 
plt.grid(True) 
plt.show
```

```python
from sklearn import metrics
print('Mean Absolute Error',metrics.mean_absolute_error(Kpgvkumamoto0414,Ypgvkumamoto0414))
print('Mean Squared Error:', metrics.mean_squared_error(Kpgvkumamoto0414,Ypgvkumamoto0414))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Kpgvkumamoto0414, Ypgvkumamoto0414)))
```

```python
from sklearn import metrics
print('Mean Absolute Error',metrics.mean_absolute_error(Kpgvkumamoto0414,YYpgvkumamoto0414))
print('Mean Squared Error:', metrics.mean_squared_error(Kpgvkumamoto0414,YYpgvkumamoto0414))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Kpgvkumamoto0414, YYpgvkumamoto0414)))
```
